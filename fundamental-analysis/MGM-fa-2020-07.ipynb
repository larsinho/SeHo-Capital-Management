## 1. Is the business simple and understandable (background)

- MGM Resorts operates integrated casino, hotel, and entertainment resorts across the United States and in Macau
- They are well positioned to take adavantage of legalization of sports betting [(link)](https://seekingalpha.com/article/4314958-top-stocks-to-buy-sports-betting-is-legalized)
- They have  about one fourth of all units in the Las Vegas
- They have a significant amount of indebtedness maturing in 2022
- They are required to pay a significant portion of our cash flows as rent.
- They rely heavily on the ability of our resorts to generate operating cash flow to fund capital expenditures, provide excess cash
flow for future development, acquisitions or investments, and repay debt financings.
- The results of operations do not tend to be seasonal in nature as all of our casino resorts operate 24 hours a day
- Stock value = book value
- They face significant competition with respect to destination travel locations generally and with respect to our peers in the
industries in which we compete.
- The business is particularly sensitive to reductions in discretionary consumer spending and corporate
spending on conventions, trade shows and business development.

### Price comparision

```python
start =  int(time.mktime(datetime.date(2000,1,1).timetuple()))
end =  int(time.mktime(datetime.date(2020,6,30).timetuple()))

# List of assets:
links =[]
ticker_list = []

# MGM Resorts International (MGM)
ticker_list.append('MGM')
# Caesars Entertainment Corporation (CZR)
ticker_list.append('CZR')
# Las Vegas Sands Corp. (LVS)
ticker_list.append('LVS')
# Hilton Worldwide Holdings Inc. (HLT)
ticker_list.append('HLT')

for ticker in ticker_list:
    links.append(myFunctions.yahooLink(start, end, ticker))

# Pulling data or loading if already pulled:
df_full = pd.DataFrame()

if os.path.isfile('./data/tourismAnalysis.pkl'):
    df_full = pd.read_pickle("./data/tourismAnalysis.pkl")
else:
    for idx, link in enumerate(links):
        print(idx, link)
        df_temp = pd.read_csv(link)
        df_temp['ticker'] = ticker_list[idx]
        df_full = df_full.append(df_temp) 
    df_full.to_pickle("./data/tourismAnalysis.pkl")    
    
```

```python
data = df_full[['Date', 'ticker', 'Adj Close']]
data.columns = ['date', 'ticker', 'adj_close']
data = data.drop_duplicates(keep = 'first')
# Displaying data
df = data.set_index('date')
table = df.pivot(columns='ticker')
#table.columns =  ['Bonds', 'Agriculture', 'Emerging', 'Europe', 'Japan', 'US', 'China', 'Gold', 'Oil'] #table.columns = [col[1] for col in table.columns]
table = table.dropna()
#table.head()
```

```python
returns = np.log(table) - np.log(table.shift(1))
table_norm = table / table.iloc[0,:] # Normalize values to start at 1
ax = plt.gca()
palette = plt.get_cmap('Set1')

styles = ['-', '--', '-.', ':']

for idx, series in enumerate(returns.mean().sort_values(ascending = False).index):
    table_norm[series].plot(figsize=(14, 7), lw=2, alpha=0.8, color=palette(idx), style = styles[idx % 4], ax=ax)
plt.legend(loc='upper left', fontsize=12)
plt.xlabel("")
plt.ylabel('Normalized prices')
```

## 5. Does management resist the institutional imperative?

5.1. An institution resists any change in its current direction

5.2. Just as work expands to fill available time, corporate projects or acquisitions will materialize to soak up available funds

5.3. Any business craving of the leader, however foolish, will quickly be supported by detailed rate-of-return and strategic studies prepared by his troops

5.4.The behaviour of peer companies, whether they are expanding, acquiring, setting executive compensation or whatever, will be mindlessly imitated

Their strategy is to increase Profit Growth Plan under the one company, one culture vision. 
- MGM 2020 will be a company-wide, business optimization initiative aimed to leverage a more centralized organization to maximize profitability and, through key investments in technology, lay the groundwork for the companys digital transformation to drive revenue growth.
- MGM Resorts is generating significant free cash flow. Over the next few years, the company expects to reallocate a portion of its annual capital expenditure budget to specific technology advancements that will increase revenues and grow market share by innovating and elevating the guest experience through data, pricing, digital and loyalty capabilities and optimizing business mix [(source)](https://www.hotelbusiness.com/mgm-resorts-unveils-strategic-plan-for-future-growth/).

It seems to be focusing on increasing revenue from existing assets, which seems like reasonable strategy. They also seek to turn a set of gaming assets into a powerful entertainment brand
